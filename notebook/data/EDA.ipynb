We can see that our data is imbalanced

```python
from sklearn.utils import resample
# create data frame minority and majority class
majority = df[(df['default.payment.next.month']==0)]
minority = df[(df['default.payment.next.month']==1)]

minority_upsample = resample(minority,
                           replace=True,
                            n_samples=23364,
                            random_state=0)

# combine majority class and unsample minority class
df = pd.concat([minority_upsample, majority])
```

```python
# Distribution of default.payment.next.month
plt.figure(figsize=(10, 5))
sns.countplot(x='default.payment.next.month', data=df)
plt.title('Distribution of default payment next month')
plt.show()
```

### Outlier Detection

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Adjust the figure size
plt.figure(figsize=(15, 30))

# Enumerate through each feature column in the DataFrame
for i, feature in enumerate(df.columns):
    plt.subplot(10, 3, i + 1)
    if df[feature].dtype == 'object':
        sns.boxplot(x=df[feature])
    else:
        sns.boxplot(x=df[feature])
    plt.title(feature)

plt.tight_layout()
plt.show()
```

We can see that our maximum columns have outlier

```python
# Impute outliers with median
def impute_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    median = df[column].median()
    df[column] = np.where((df[column] < lower_bound) | (df[column] > upper_bound), median, df[column])
    return df

for col in df.select_dtypes(include=['float64', 'int64']).columns:
    df = impute_outliers(df, col)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Adjust the figure size
plt.figure(figsize=(15, 30))

# Enumerate through each feature column in the DataFrame
for i, feature in enumerate(df.columns):
    plt.subplot(10, 3, i + 1)
    if df[feature].dtype == 'object':
        sns.boxplot(x=df[feature])
    else:
        sns.boxplot(x=df[feature])
    plt.title(feature)

plt.tight_layout()
plt.show()
```
